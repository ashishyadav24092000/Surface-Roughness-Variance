## Method-2 -----> Using the Python tools :

```python
import pandas as pd 
import numpy as np
import math
from scipy import stats
```

## Step-1 ---> Formulate the hypothesis.

H0 : 
    σ1_square = σ2_square(there exist a difference between the two variances)




Ha :
    σ1_square ≠ σ2_square(there is no difference between the two variances)

## Step-2 ----> Calculate degree of freedom for both the samples.

```python
sample_1_len = 11
numerator_dof = (sample_1_len)-1

sample_2_len = 16
denominator_dof = (sample_2_len)-1

print('Numerator degree of freedom is : {} \nDenominator degree of freedom is : {}'.format(numerator_dof,denominator_dof))
```

## Step-3 -----> Calculate upper limit and lower limit for the f-test at α = 0.10/2

```python
import scipy
lower_limit = scipy.stats.f.ppf(q=0.05,dfn=15,dfd=10)

upper_limit = scipy.stats.f.ppf(q=1-0.05,dfn=15,dfd=10)

print('The upper limit is : {} \nThe lower limit is: {}'.format(lower_limit,upper_limit))
```

## Since this interval has the presence of unity i.e. 1. So we do not reject our null Hypothesis.

# Final Insight /  Outcome ---> the standard deviation for the roughness of surfaces for the two processes cannot be said to be different from one another with an significance level 0f 10 percent.
